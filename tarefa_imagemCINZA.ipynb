```python
from google.colab import files

# Fazer upload da imagem
uploaded = files.upload()

# Pega o nome do arquivo enviado
image_name = list(uploaded.keys())[0]
print(f"Imagem carregada: {image_name}")

```

```python
import numpy as np
import cv2
from PIL import Image

# Abrir a imagem e converter para escala de cinza
image = Image.open(image_name)  # Abre a imagem original
gray_image = image.convert("L")  # Converte para escala de cinza

# Converte para um array numérico (matriz de pixels)
image_array = np.array(gray_image)

# Exibe a imagem em tons de cinza
gray_image.show()
```

```python
# Criar imagem preto e branco (binarizada)
threshold = 128  # Limite para binarização
binary_image_array = np.where(image_array > threshold, 255, 0)  # Converte para 0 ou 255

# Converter de volta para imagem
binary_image = Image.fromarray(np.uint8(binary_image_array))

# Exibir a imagem preto e branco
binary_image.show()
```

```python
import matplotlib.pyplot as plt

# Criando a figura com duas imagens
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrando a imagem em tons de cinza
ax[0].imshow(gray_image, cmap="gray")
ax[0].set_title("Imagem em Cinza")
ax[0].axis("off")  # Oculta os eixos

# Mostrando a imagem preto e branco
ax[1].imshow(binary_image, cmap="gray")
ax[1].set_title("Imagem Preto e Branco")
ax[1].axis("off")

plt.show()
```
